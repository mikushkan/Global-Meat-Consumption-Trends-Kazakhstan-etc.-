```python
library(readr)

meat_data <- read_csv(
  "/kaggle/input/datasets/mikushkan/global-meat-consumption-trends-kazakhstan/per-capita-meat-consumption-by-type-kilograms-per-year.csv"
)

head(meat_data)
library(tidyverse)
glimpse(meat_data)
summary(meat_data)
library(tidyverse)
library(readr)
library(ggcorrplot)
library(forcats)
library(gganimate)

meat_data <- read_csv(
  "/kaggle/input/datasets/mikushkan/global-meat-consumption-trends-kazakhstan/per-capita-meat-consumption-by-type-kilograms-per-year.csv",
  show_col_types = FALSE
)

colnames(meat_data) <- c(
  "Country",
  "Code",
  "Year",
  "Other",
  "SheepGoat",
  "BeefBuffalo",
  "Pig",
  "Poultry"
)

head(meat_data)

meat_data %>%
  filter(Country == "Kazakhstan") %>%
  pivot_longer(
    cols = Other:Poultry,
    names_to = "MeatType",
    values_to = "KgPerCapita"
  ) %>%
  ggplot(aes(x = Year, y = KgPerCapita, color = MeatType)) +
  geom_line(linewidth = 1) +
  theme_minimal()

# Суммарное потребление мяса

meat_summary <- meat_data %>%
  mutate(
    TotalMeat = Other + SheepGoat + BeefBuffalo + Pig + Poultry
  ) %>%
  filter(
    Country %in% c(
      "Kazakhstan",
      "Russia",
      "China",
      "United States",
      "Brazil"
    )
  ) %>%
  filter(!is.na(TotalMeat))
         
         ggplot(
  meat_summary,
  aes(
    x = Year,
    y = TotalMeat,
    color = Country
  )
) +
  geom_line(linewidth = 1.2) +
  labs(
    title = "Суммарное потребление мяса на душу населения",
    x = "Год",
    y = "Кг мяса на человека",
    color = "Страна"
  ) +
  theme_minimal()

  # Доли видов мяса в Казахстане

kz_data <- meat_data %>%
  filter(Country == "Kazakhstan") %>%
  select(
    Year,
    Other,
    SheepGoat,
    BeefBuffalo,
    Pig,
    Poultry
  ) %>%
  na.omit()

  kz_long <- kz_data %>%
  pivot_longer(
    cols = -Year,
    names_to = "MeatType",
    values_to = "KgPerCapita"
  )

  kz_long <- kz_long %>%
  group_by(Year) %>%
  mutate(
    Share = KgPerCapita / sum(KgPerCapita)
  )

  ggplot(
  kz_long,
  aes(
    x = Year,
    y = Share,
    fill = MeatType
  )
) +
  geom_area(
    alpha = 0.8,
    linewidth = 0.5,
    colour = "white"
  ) +
  scale_y_continuous(labels = scales::percent) +
  labs(
    title = "Доли видов мяса в Казахстане",
    x = "Год",
    y = "Доля (%)",
    fill = "Вид мяса"
  ) +
  theme_minimal()

  kz_pie <- meat_data %>%
  filter(
    Country == "Kazakhstan",
    Year == 2020
  ) %>%
  select(
    Other,
    SheepGoat,
    BeefBuffalo,
    Pig,
    Poultry
  ) %>%
  pivot_longer(
    cols = everything(),
    names_to = "MeatType",
    values_to = "KgPerCapita"
  ) %>%
  mutate(
    Percent = KgPerCapita / sum(KgPerCapita) * 100
  )

  ggplot(
  kz_pie,
  aes(
    x = " ",
    y = Percent,
    fill = MeatType
  )
) +
  geom_col(width = 1) +
  coord_polar("y") +
  labs(
    title = "Структура потребления мяса в Казахстане (2020)"
  ) +
  theme_void() +
  theme(legend.title = element_blank())

  years <- c(1990, 2000, 2010, 2020)

kz_bar <- meat_data %>%
  filter(
    Country == "Kazakhstan",
    Year %in% years
  ) %>%
  select(
    Year,
    Other,
    SheepGoat,
    BeefBuffalo,
    Pig,
    Poultry
  ) %>%
  pivot_longer(
    cols = -Year,
    names_to = "MeatType",
    values_to = "KgPerCapita"
  ) %>%
  group_by(Year) %>%
  mutate(
    Share = KgPerCapita / sum(KgPerCapita)
  )

ggplot(
  kz_bar,
  aes(
    x = factor(Year),
    y = Share,
    fill = MeatType
  )
) +
  geom_bar(stat = "identity") +
  scale_y_continuous(labels = scales::percent) +
  labs(
    title = "Изменение структуры потребления мяса",
    x = "Год",
    y = "Доля (%)",
    fill = "Вид мяса"
  ) +
  theme_minimal()

 russia_data <- meat_data %>%
  filter(Country == "Russia") %>%
  select(
    Year,
    Other,
    SheepGoat,
    BeefBuffalo,
    Pig,
    Poultry
  ) %>%
  na.omit()

 russia_data <- russia_data %>%
  mutate(
    TotalMeat = Other + SheepGoat + BeefBuffalo + Pig + Poultry
  )

 ggplot(
  russia_data,
  aes(
    x = Year,
    y = TotalMeat
  )
) +
  geom_line(
    color = "darkred",
    linewidth = 1.2
  ) +
  labs(
    title = "Потребление мяса в России",
    x = "Год",
    y = "Кг мяса на человека"
  ) +
  theme_minimal()

compare_data <- meat_data %>%
  filter(
    Country %in% c(
      "Kazakhstan",
      "Russia",
      "United States"
    )
  ) %>%
  mutate(
    TotalMeat = Other + SheepGoat + BeefBuffalo + Pig + Poultry
  ) %>%
  select(
    Country,
    Year,
    TotalMeat
  ) %>%
  filter(!is.na(TotalMeat))

ggplot(
  compare_data,
  aes(
    x = Year,
    y = TotalMeat,
    color = Country
  )
) +
  geom_line(linewidth = 1.2) +
  labs(
    title = "Сравнение потребления мяса",
    x = "Год",
    y = "Кг мяса на человека",
    color = "Страна"
  ) +
  theme_minimal()

library(forcats)

selected_countries <- c(
  "Kazakhstan",
  "Russia",
  "United States",
  "Germany",
  "Japan",
  "China",
  "Saudi Arabia"
)

heatmap_data <- meat_data %>%
  filter(Country %in% selected_countries) %>%
  mutate(
    TotalMeat = Other + SheepGoat + BeefBuffalo + Pig + Poultry
  ) %>%
  select(
    Country,
    Year,
    TotalMeat
  ) %>%
  filter(!is.na(TotalMeat))
         
ggplot(
  heatmap_data,
  aes(
    x = Year,
    y = fct_reorder(
      Country,
      TotalMeat,
      .fun = mean
    )
  )
) +
  geom_tile(aes(fill = TotalMeat)) +
  scale_fill_viridis_c(
    option = "inferno",
    name = "Кг мяса"
  ) +
  labs(
    title = "Тепловая карта потребления мяса",
    x = "Год",
    y = "Страна"
  ) +
  theme_minimal()

corr_data <- meat_data %>%
  select(
    Pig,
    Poultry
  ) %>%
  filter(
    !is.na(Pig),
    !is.na(Poultry)
  )

ggplot(
  corr_data,
  aes(
    x = Pig,
    y = Poultry
  )
) +
  geom_point(
    alpha = 0.3,
    color = "blue"
  ) +
  geom_smooth(
    method = "lm",
    color = "red"
  ) +
  labs(
    title = "Корреляция свинины и курятины",
    x = "Свинина",
    y = "Курятина"
  ) +
  theme_minimal()

library(ggcorrplot)

plot_correlation <- function(country_name) {

  data_country <- meat_data %>%
    filter(Country == country_name) %>%
    select(
      Pig,
      Poultry,
      BeefBuffalo,
      SheepGoat,
      Other
    ) %>%
    drop_na()

  corr_matrix <- cor(data_country)

  ggcorrplot(
    corr_matrix,
    method = "circle",
    type = "lower",
    lab = TRUE,
    title = paste(
      "Корреляция видов мяса —",
      country_name
    )
  )
}

plot_correlation("Kazakhstan")
plot_correlation("Russia")
plot_correlation("United States")
plot_correlation("China")
plot_correlation("Germany")
```
